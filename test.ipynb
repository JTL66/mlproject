```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

BASE_URL = 'https://www.basketball-reference.com/leagues/NBA_2018_games-{month}.html' 
all_month = np.array(['october','november','december','january','february','march','april','may','june'])

def get_content():
    list = []
    for i in range(len(all_month)):
        url = BASE_URL.format(month=all_month[i])
        print(url)
        html = urlopen(url).read()
        bsObj = BeautifulSoup(html,'lxml')
        rows = [dd for dd in bsObj.select('tbody tr')] #select the rows
        for row in rows:
            cell = [i.text for i in row.find_all('td')] #select td from rows
            list.append(cell)
    return list #back to the lsit

#save as csv file
def save():
    file = open('jtliu/Campus/20FallUNH/DSCI6003/Project/NBA_predict-master/matches.csv','w') #write the path of the file
    list = get_content()
    df = pd.DataFrame(columns=[1,2,3,4,5,6,7,8,9] ,data=list)
    df.to_csv(file)
    print('done')
```
